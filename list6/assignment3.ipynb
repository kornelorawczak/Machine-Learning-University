```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def cluster_image(im_number, k):
    path = f"p{im_number}.jpg"
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pixel_list = image.reshape(-1, 3)

    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(pixel_list)
    compressed_pixels = kmeans.cluster_centers_[labels]
    compressed_image = compressed_pixels.reshape(image.shape).astype(np.uint8)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(compressed_image)
    plt.title("Dominant Colors (K=%d)" % k)
    plt.axis('off')

    plt.show()

cluster_image(1, 3)
cluster_image(2, 3)
cluster_image(3, 3)
cluster_image(4, 3)
cluster_image(5, 3)
```
