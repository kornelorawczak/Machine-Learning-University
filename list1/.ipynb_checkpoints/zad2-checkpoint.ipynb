```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import numpy as np


data = pd.read_csv("bankmarketing/bank/bank-full.csv", sep=";")
# ---------------------------
# 1:
all_data = data["duration"]
yes_data = data[data["y"] == "yes"]["duration"]
no_data  = data[data["y"] == "no"]["duration"]

_, axes = plt.subplots(1, 3, figsize=(14, 6), sharey=True)

axes[0].hist(all_data, bins=30, color="skyblue", edgecolor="black")
axes[0].set_title("All data on calls duration")
axes[0].set_xlabel("Call duration in seconds")
axes[0].set_ylabel("no. of samples")

axes[1].hist(yes_data, bins=30, color="green", edgecolor="black")
axes[1].set_title("Calls that lead to subscription")
axes[1].set_xlabel("Call duration in seconds")
#axes[1].set_ylabel("no. of samples")

axes[2].hist(no_data, bins=30, color="red", edgecolor="black")
axes[2].set_title("Calls that didn't lead to subscription")
axes[2].set_xlabel("Call duration in seconds")
#axes[2].set_ylabel("no. of samples")

plt.tight_layout()
plt.show()
```

```python
# 2:
def create_plot(minAge):
    filtered_data = data[data["age"] > minAge]
    plt.figure(figsize=(8,4))
    plt.hist(filtered_data['balance'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f"Balance of people over {minAge} years old")
    plt.xlabel("Balance")
    plt.ylabel("No. of people")
    plt.grid(alpha=0.2)
    plt.show()

interact(create_plot, minAge=IntSlider(value=25, min=18, max=100, step=1, description='Min Age'))
```

```python
# 3:
thresholds = np.arange(0, 5000, 100)
positive_ratios = []
for time in thresholds: 
    filtered_data = data[data["duration"] > time]
    if len(filtered_data) > 0:
        yes_data = filtered_data[filtered_data["y"] == "yes"]["duration"]
        positive_ratios.append(100 * yes_data.count() / filtered_data["y"].count())
plt.figure(figsize=(8,5))
plt.plot(thresholds, positive_ratios, marker='o', color='royalblue')
plt.title("Percent of positive calls with the duration > t")
plt.xlabel("Time in seconds")
plt.ylabel("% of positive calls")
plt.grid(alpha=0.3)
plt.show()
        
```

```python
# 4:
thresholds = np.arange(-9000, 105000, 5000)
positive_ratios = []
for b in thresholds: 
    filtered_data = data[data["balance"] > b]
    if len(filtered_data) > 0:
        yes_data = filtered_data[filtered_data["y"] == "yes"]["balance"]
        positive_ratios.append(100 * yes_data.count() / filtered_data["y"].count())

overall_ratio = 100 * data[data["y"] == "yes"]["balance"].count() / data["y"].count()

plt.figure(figsize=(8,5))
plt.plot(thresholds, positive_ratios, marker='o', color='royalblue')
plt.axhline(y=overall_ratio, color='red', linestyle='--', label=f'random campaign ({overall_ratio:.2f}%)')
plt.title("Percent of positive calls with the balance > b")
plt.xlabel("Balance")
plt.ylabel("% of positive calls")
plt.grid(alpha=0.3)
plt.show()
```

```python
n_total = len(data)
yes_total = len(data[data["y"] == "yes"])
part1 = data.sample(frac=0.5, random_state=42) #random_state=42 ensures that the results are reproductible
part2 = data.drop(part1.index)
p1_total = len(part1)
p1_yes = len(part1[part1["y"] == "yes"])
p2_total = len(part2)
p2_yes = len(part2[part2["y"] == "yes"])

print(f' All the people: {n_total} \n All people who signed: {yes_total} \n No. of people in part1: {p1_total} \n No. of people who signed in part1: {p1_yes} \n No. of people in part2: {p2_total} \n No. of people who signed in part2: {p2_yes}')

#percentyle = np.linspace(0, 100, 50) 
#thresholds = np.unique(np.percentile(data['balance'], percentyle))
#thresholds = np.sort(thresholds)  
thresholds = np.arange(-9000, 105000, 5000)
rows = []
for b in thresholds:
    subset = part1[part1['balance'] > b]
    if len(subset) == 0:
        continue

    yesPercentage = len(subset[subset["y"] == "yes"]) / len(subset) if len(subset) != 0 else 0
    yesSelected = len(subset[subset['y'] == 'yes']) / len(part1[part1['y'] == 'yes'])
    Hm = 2 * yesPercentage * yesSelected / (yesPercentage + yesSelected) if (yesPercentage + yesSelected) > 0 else 0
    rows.append({'threshold': b, 'hm': Hm})

res = pd.DataFrame(rows)
def calculate_errors(d, th):
    predicted_positive = d[d["balance"] > th]
    predicted_negative = d[d["balance"] <= th]
    n = len(data)
    positives = len(d[d["y"] == "yes"])
    
    # False Positive: people above the threshold that said no
    false_pos = len(predicted_positive[predicted_positive["y"] == "no"])
    false_pos_rate = 100 * false_pos / len(predicted_positive) if len(predicted_positive) != 0 else 0
    
    # False Negative: people <= threshold that would've said yes
    false_neg = len(predicted_negative[predicted_negative["y"] == "yes"])
    false_neg_rate = 100 * false_neg / positives if positives != 0 else 0
    
    return false_pos_rate, false_neg_rate

# Threshold 1: 90th percentile (top 10% of balances)
t1 = np.percentile(data['balance'], 90)
# Threshold 2: Maximum harmonic mean on the first part 
t2 = res.loc[res['hm'].idxmax(), 'threshold']
# Threshold 3: Median 
t3 = data['balance'].median()
# Threshold 4: Finding the best threshold using brute force
balances = np.sort(data['balance'].unique())
t4 = 0
best_error = float('inf')

for t in range(balances[0], balances[-1], 100):
    fp_rate, fn_rate = calculate_errors(part1, t)
    total_error = fp_rate + fn_rate  
    if total_error < best_error:
        best_error = total_error
        t4 = t

generalThresholds = [t1, t2, t3, t4]
labels = ['90th percentile', 'max harmonic mean', 'median', 'best (xd)']


results = []
for name, part in [("part1", part1), ("part2", part2)]:
    for t, label in zip(generalThresholds, labels):
        fp_rate, fn_rate = calculate_errors(part, t)
        results.append({
            "part": name,
            "threshold": label,
            "false_positive_pct": fp_rate,
            "false_negative_pct": fn_rate
        })

res_data = pd.DataFrame(results)
_, axes = plt.subplots(1, 4, figsize=(16,5), sharey=True)

for i, (t, label) in enumerate(zip(generalThresholds, labels)):
    subset = res_data[res_data["threshold"] == label]
    
    axes[i].bar(subset["part"], subset["false_positive_pct"], label="False Positives", alpha=0.7)
    axes[i].bar(subset["part"], subset["false_negative_pct"], bottom=subset["false_positive_pct"], 
                label="False Negatives", alpha=0.7)
    
    axes[i].set_title(f"Threshold {label}", fontsize=12)
    axes[i].set_ylabel("Error percentage")
    axes[i].set_xlabel("data part")
    axes[i].grid(alpha=0.2)
    axes[i].legend(fontsize=8)

plt.suptitle("Comparing errors between 4 thresholds and 2 parts of data", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
