```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

data = np.random.randn(1_000_000)

_ = plt.hist(data, bins=100, density=True, alpha=0.8, color="blue")
x = np.linspace(-5, 5, 1000)
n_data = norm.pdf(x, 0, 1)

plt.plot(x, n_data, label="Density function for normal distribution", color="red")
plt.title("1,000,000 probes generated from normal distribution vs real density function")
plt.legend()
plt.grid(True)
plt.show()
```

```python
s = 1

data = pd.read_csv("iris/iris.data", header=None)
data.columns = ["s-len", "s-width", "p-len", "p-width", "class"]

np.random.seed(42) # to make it repetetive throught launches
noised_data = data.copy()
noised_data["s-len"]   += np.random.normal(0, s, size=len(noised_data))
noised_data["s-width"] += np.random.normal(0, s, size=len(noised_data))

types_colors = {
    "Iris-setosa" : "blue",
    "Iris-versicolor" : "red",
    "Iris-virginica" : "green"
}

_, axes = plt.subplots(1, 2, figsize=(14, 6), sharex=True, sharey=True)

for kind, color in types_colors.items():
    subset = data[data["class"] == kind]
    axes[0].scatter(
        subset["s-len"],           # X
        subset["s-width"],         # Y
        s=subset["p-len"] * 30,    # size
        color=color,
        alpha=0.8,
        label=kind
    )

axes[0].set_title("Data without noise")
axes[0].set_xlabel("Sepal Length")
axes[0].set_ylabel("Sepal Width")
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.6)

for kind, color in types_colors.items():
    subset = noised_data[noised_data["class"] == kind]
    axes[1].scatter(
        subset["s-len"], 
        subset["s-width"], 
        s=subset["p-len"] * 30,
        color=color,
        alpha=0.8,
        label=kind
    )
axes[1].set_title(f"Data with noise from N(0, {s})")
axes[1].set_xlabel("Sepal Length")
axes[1].grid(True, linestyle='--', alpha=0.6)
axes[1].legend()

plt.suptitle("Relationship between iris type and its sepal length (X), width (Y) and petal length (size)", fontsize=14)
plt.tight_layout()
plt.show()
```
